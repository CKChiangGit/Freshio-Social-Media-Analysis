```python
df = pd.read_csv('locations.csv')
df
```

```python
# Filter out rows that have non-null businessZipCode
zip_code_mask = df['businessZipCode'].notna()

# Convert those to int, then to string, and then pad with zeros
df.loc[zip_code_mask, 'businessZipCode'] = df.loc[zip_code_mask, 'businessZipCode'].astype(int).astype(str).str.zfill(6)

df
```

```python
import pandas as pd
from geopy.geocoders import Nominatim
import time
from random import randint

# Create a geolocator function to get lat and lon from postal codes
def get_lat_lon(postal_code, country="Singapore"):
    user_agent1 = 'user_me_{}'.format(randint(1,99999))
    geolocator = Nominatim(user_agent=user_agent1)
    location = geolocator.geocode(f"{postal_code}, {country}")
    if location:
        return (location.latitude, location.longitude)
    else:
        return (None, None)

# Get latitudes and longitudes for each postal code in the DataFrame
latitudes = []
longitudes = []

for postal_code in df['businessZipCode']:
    lat, lon = get_lat_lon(postal_code)
    latitudes.append(lat)
    longitudes.append(lon)
    time.sleep(1)  # To avoid hitting rate limits

df['Latitude'] = latitudes
df['Longitude'] = longitudes
```

```python
import folium

# Assuming you have the dataframe loaded as 'df'
m2 = folium.Map(location=[1.3521, 103.8198], zoom_start=12)  # Centered on Singapore

for index, row in df.iterrows():
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup=f"<strong>{row['fullName']}</strong><br>{row['profileUrl']}",
        icon=folium.Icon(icon="cloud"),
    ).add_to(m2)

# Display the map
m2
```
